### IGNORING WARNINGS

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### IMPORTING AND CHECKING DATASET

```python
import pandas as pd
df = pd.read_csv('train_mobile.csv')
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### SEPERATING X AND Y FEATURES

```python
x = df.drop(columns=['price_range'])
y = df[['price_range']]
```

```python
x.head()
```

```python
y.head()
```

### PREPROCESSING THE DATA ON X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
x_pre = num_pipe.fit_transform(x)
x_pre.head()
```

### APPLYING TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x_pre, y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtest.shape[1]
```

### USING ONEHOTENCODER FOR MULTICLASS CLASSIFICATION

```python
from keras.utils import to_categorical
ytrain_cat = to_categorical(ytrain)
ytest_cat = to_categorical(ytest)
```

```python
ytrain_cat
```

```python
ytest_cat
```

```python
ytest.head()
```

```python
df['price_range'].unique()
```

### BUILDING NEURAL NETWORK

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
```

```python
model = Sequential()

# CREATING DENSE LAYER 1
model.add(Dense(units=64, activation='relu', input_shape=(xtrain.shape[1],)))

# ADDING DROPOUT LAYER 1
model.add(Dropout(rate=0.3))

# CREATING DENSE LAYER 2
model.add(Dense(units=32, activation='relu'))

# ADDING DROPOUT LAYER 2
model.add(Dropout(rate=0.2))

# OUTPUT LAYER SHOULD HAVE 4 NEURONS BECAUSE THERE ARE 4 CLASSES IN THIS DATA
model.add(Dense(units=4, activation='softmax'))
```

```python
model.summary()
```

### COMPILING THE MODEL

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
es = EarlyStopping(monitor='val_loss', patience=10)
```

### TRAINING THE MODEL

```python
hist = model.fit(xtrain, ytrain_cat, validation_split=0.2,
epochs=200, callbacks=[es])
```

### PLOTTING THE CURVE

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(hist.history['loss'], label='TRAINING LOSS')
plt.plot(hist.history['val_loss'], label='VALIDATION LOSS')
plt.xlabel('EPOCHS')
plt.ylabel('CATEGORICAL CROSS ENTROPY')
plt.title('LEARNING CURVE')
plt.legend()
plt.show()
```

### EVALUATING THE MODEL

```python
model.evaluate(xtrain, ytrain_cat)
```

```python
model.evaluate(xtest, ytest_cat)
```

### GETTING PROBABILITY FOR TRAIN AND TEST

```python
yprob_train = model.predict(xtrain)
yprob_test = model.predict(xtest)
```

```python
yprob_test[0:5]
```

```python
yprob_train[0:5]
```

### GETTING PREDICTIONS OUT OF PROBABILITY

```python
import numpy as np
ypred_train = np.argmax(yprob_train, axis=1)
ypred_test = np.argmax(yprob_test, axis=1)
```

```python
ypred_train[0:5]
```

```python
ytrain.head()
```

```python
ypred_test[0:5]
```

```python
ytest.head()
```

### PLOTTING THE CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_test)
cfd = ConfusionMatrixDisplay(cf)
cfd.plot()
```

### PRINTING CLASSIFICATION REPORT

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```
